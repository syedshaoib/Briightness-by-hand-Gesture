```python
import cv2
import mediapipe as mp
import pyttsx3
import numpy as np
import math
import screen_brightness_control as sbc

 
#set the brightness of the primary display to 75%

 
print(sbc.get_brightness())


# define a video capture object
vid = cv2.VideoCapture(0)
myhands=mp.solutions.hands
hands=myhands.Hands()
cx=0
cy=0
cy1=0
cx1=0

mpDraw=mp.solutions.drawing_utils
while True:
 
    success, frame = vid.read()
    frame=cv2.flip(frame,1)
    

    imageRGB=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    results = hands.process(imageRGB)
    if results.multi_hand_landmarks:
    
       
        for handsA in results.multi_hand_landmarks:
          
            mpDraw.draw_landmarks(frame,handsA,myhands.HAND_CONNECTIONS)
            for id,lm in enumerate(handsA.landmark):
               
                if(id==8):
                    cx1=int(frame.shape[1] *lm.x)
                    cy1=int(frame.shape[0] * lm.y)
                    cv2.circle(frame,(cx1,cy1),5,(115,66,133),9)
                   
                if(id==4):
                    cx=int(frame.shape[1] *lm.x)
                    cy=int(frame.shape[0] * lm.y)
                    cv2.circle(frame,(cx,cy),5,(115,66,133),9)
#                 cv2.line(frame,(cx1,cy1),(cx,cy),(33,66,77),3)
                length=math.hypot(cx-cx1,cy-cy1)
                
                vol= np.interp(length,[-8,100],[0,100])
                sbc.set_brightness(int(vol), display=0)
                print(vol)
                
            
                
             
            
                    
                  
                
            
                
            
            
            
    cv2.imshow('frame', frame)
    

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
vid.release()
# Destroy all the windows
cv2.destroyAllWindows()
```
